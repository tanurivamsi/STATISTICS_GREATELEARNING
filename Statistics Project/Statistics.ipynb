1. Read the dataset and replace dashes with 0 (2.5 points)

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Laliga.csv')
# Assign new column names to the DataFrame
data.columns = ['Pos', 'Team', 'Seasons', 'Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 'GamesLost', 
                'GoalsFor', 'GoalsAgainst', 'Champion', 'Runner-up', 'Third', 'Fourth', 'Fifth', 'Sixth', 
                'T', 'Debut', 'Since/LastApp', 'BestPosition']

# Delete the second row
data = data.drop(1)

# Reset index after dropping the row (optional, to keep index sequential)
data.reset_index(drop=True, inplace=True)

# Replace dashes with 0
data.replace('-', 0, inplace=True)

# Confirm replacement
print(data.head())
```

2. Print all the teams that started playing between 1930-1980 (5 points)

```python
data.head(3)
```

```python
# Convert 'Year' to integer if necessary
data['Debut'] = pd.to_numeric(data['Debut'], errors='coerce')

# Filter teams that started between 1930 and 1980
teams_1930_1980 = data[(data['Debut'] >= 1930) & (data['Debut'] <= 1980)]

# Print the filtered teams
print(teams_1930_1980[['Team', 'Debut']])
```

3. Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
# Convert 'Points' column to numeric, replacing non-numeric values with NaN
data['Points'] = pd.to_numeric(data['Points'], errors='coerce')

# Replace NaN values in 'Points' with 0 (or any appropriate value)
data['Points'].fillna(0, inplace=True)

# Sort teams by 'Points' and get the top 5
top_5_teams = data.sort_values(by='Points', ascending=False).head(5)

# Print top 5 teams
print(top_5_teams[['Team', 'Points']])

```

4. Write a function Goal_diff_count to return teams with their goal differences (5 points)

```python
data.head(3)
```

```python
def Goal_diff_count(df):
    # Calculate goal difference
    df['Goal Difference'] = pd.to_numeric(df['GoalsFor'], errors='coerce') - pd.to_numeric(df['GoalsAgainst'], errors='coerce')
    
    # Find the team with max and min goal difference
    max_diff_team = df[df['Goal Difference'] == df['Goal Difference'].max()]['Team'].values[0]
    min_diff_team = df[df['Goal Difference'] == df['Goal Difference'].min()]['Team'].values[0]
    
    return df[['Team', 'Goal Difference']], max_diff_team, min_diff_team

# Call the function
goal_diff_df, max_team, min_team = Goal_diff_count(data)

# Print the teams with their goal differences
print(goal_diff_df)

# Print the teams with max and min goal difference
print(f"Team with maximum goal difference: {max_team}")
print(f"Team with minimum goal difference: {min_team}")
```

5. Create a new column 'Winning Percent' (5 points)

```python
data.head(3)
```

```python
# Calculate Winning Percentage
data['Winning Percent'] = (pd.to_numeric(data['GamesWon'], errors='coerce') / pd.to_numeric(data['GamesPlayed'], errors='coerce')) * 100

# Replace NaN or infinite values with 0%
data['Winning Percent'].fillna(0, inplace=True)
data['Winning Percent'].replace([float('inf'), -float('inf')], 0, inplace=True)

# Print the top 5 teams with highest Winning Percent
top_5_winning_percent = data.sort_values(by='Winning Percent', ascending=False).head(5)
print(top_5_winning_percent[['Team', 'Winning Percent']])
```

```python
# Group by 'Best Position' and sum the points
grouped_by_position = data.groupby('BestPosition')['Points'].sum()

# Print the grouped data
print(grouped_by_position)
```
